```python
from sklearn import datasets 
from sklearn.linear_model import LinearRegression 
import matplotlib.pyplot as plt 

# Load data, inspect and to do exploratory plots 
#  
diabetes = datasets.load_diabetes() 

X = diabetes.data 
t = diabetes.target 

#Inspect Size 
# 
NumData, NumFeatures = X.shape 

print(NumData, NumFeatures) # 442*10 
print(t.shape)              # 442 
```

```python
# Plot and Save 
# 
fig,ax = plt.subplots(nrows = 1, ncols = 2, figsize = (12,4)) 
ax[0].hist(t, bins = 40) 
ax[1].scatter(X[:,6], X[:,7], c = 'm', s = 3) 
ax[1].grid(True) 
plt.tight_layout() 
plt.savefig("DiabetsTargetAndTwoInputs1.jpg") 
```

```python
import numpy as np 
#Linear regression using sklearn 
#Plot Prediction to check if they look hte same 
lin = LinearRegression() 
lin.fit(X,t) 
th1 = lin.predict(X) 
plt.scatter(t, th1, c = 'm', s =3) 
plt.savefig("LinearRgerssion From Sklearn1.png")
D1 = th1 - t
Difference1 = np.linalg.norm(D1)
print(Difference1)
```

```python
#Pseudo-increase solution to linear regression 
#Plot Prediction to check if they look hte same 
w = np.linalg.pinv(X.T @ X) @ X.T @ t 
print(w)
th2 = X @ w 
plt.scatter(t, th2, c = 'm', s =3) 
plt.savefig("Pseudo-Inverse1.png")
D2 = th2 - t
Difference2 = np.linalg.norm(D2)
print(Difference2)
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import lasso_path
from sklearn import datasets

# synthetic daata:
# problem taken from Hastle, et al., Statistical Learning with Sparsity
# Z1, Z2 ~ N(0,1)
# Y = 3*Z1 -1.5*Z2 + 10*N(0,1) Noisy response
# Noisy input (the six are in two groups of three each)
# Xj = Z1 + 0.2*N(0,1) for j = 1,2,3, and
# Xj = Z2 + 0.2*N(0,1) for j = 4,5,6.

N = 100
y0 = np.empty(0)
X0 = np.empty([0,6])
for i in range (N):
    Z1 = np.random.randn()
    Z2 = np.random.randn()
    y0 = np.append(y0, 3*Z1 -1.5*Z2 + 10*np.random.randn())
    Xarr0 = np.array([Z1,Z1,Z1,Z2,Z2,Z2]) + np.random.randn(6)/5
    X0 = np.vstack ((X0, Xarr0.tolist()))
```

```python
# Compute regressions with Lasso and return paths
#
alphas_lasso, coefs_lasso, _ = lasso_path(X0, y0, fit_intercept = False)

# Plot each coefficient
#
fig, ax = plt.subplots(figsize = (8,4))
for i in range(6):
    ax.plot(alphas_lasso, coefs_lasso[i,:])
    
ax.grid(True)
ax.set_xlabel("Regularization Hyperparameters")
ax.set_ylabel("Regression Coefficients")
```

```python
import pandas as pd
sol = pd.read_excel("Husskonen_Solubility_Features.xlsx", verbose = False).dropna()
print(sol.shape)

colnames = sol.columns
print(colnames)

t = sol["LogS.M."].to_numpy()
X = sol[colnames[5:len(colnames)-1]].to_numpy()
N, p = X.shape

# Splir data into training and test sets
#
from sklearn.model_selection import train_test_split
X_train, X_test, t_train, t_test = train_test_split(X,t,test_size=0.3)

#Regularized regression
#
gamma = 0.1
w = np.linalg.inv(X_train.T @ X_train + gamma*np.identity(p)) @ X_train.T @t_train
th_train = X_train @ w
th_test = X_test @ w

#Plot training and test predictions
#
fig,ax = plt.subplots(nrows = 1, ncols = 2, figsize = (10,4))
ax[0].scatter(t_train, th_train, c='m', s=3)
ax[0].grid(True)
ax[0].set_title("Training Data", fontsize =14)

fig,ax = plt.subplots(nrows = 1, ncols = 2, figsize = (10,4))
ax[1].scatter(t_test, th_test, c='m', s=3)
ax[1].grid(True)
ax[1].set_title("Test Data", fontsize =14)

plt.savefig("SolubilityPrediction.jpg")
```

```python
#Over to you implement Lasso
#
lx = Lasso(alpha = 0.2,max_iter = 10000000)
lx.fit(X_train,t_train)
th_train_lasso = lx.predict(X_train)
th_test_lasso = lx.predict(X_test)

fig,ax = plt.subplots(nrows = 1, ncols = 2, figsize = (10,4))
ax[0].scatter(t_train, th_train_lasso, c='m', s=3)
ax[0].grid(True)
ax[0].set_title("Training Data", fontsize =14)

fig,ax = plt.subplots(nrows = 1, ncols = 2, figsize = (10,4))
ax[1].scatter(t_test, th_test_lasso, c='m', s=3)
ax[1].grid(True)
ax[1].set_title("Test Data", fontsize =14)

plt.savefig("SolubilityPredictionLasso.jpg")
```
